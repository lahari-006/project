```python
def recursive_chunking(text, max_len=50):
    if len(text) <= max_len:
        return [text]
    mid = len(text)//2
    return recursive_chunking(text[:mid], max_len) + recursive_chunking(text[mid:], max_len)

text = """Climate change is one of the biggest challenges of our time.
It affects weather patterns, sea levels, and biodiversity.

Governments around the world are working on policies to reduce carbon emissions.
Renewable energy sources like solar and wind are becoming more popular.

Individuals can also make a difference by reducing waste, conserving energy, and supporting sustainable practices."""
chunks = recursive_chunking(text, 25)
print(chunks)
```

```python
def parent_child_chunking(text, parent_size=50, child_size=20):
    parents = [text[i:i+parent_size] for i in range(0, len(text), parent_size)]
    hierarchy = []
    for parent in parents:
        children = [parent[j:j+child_size] for j in range(0, len(parent), child_size)]
        hierarchy.append({"parent": parent, "children": children})
    return hierarchy

text = "This is a long text that will be split into parent and child chunks."
chunks = parent_child_chunking(text, 40, 15)
print(chunks)
```

```python
def late_chunking(text, chunk_size=30):
    # Step 1: Process the text first (clean/normalize)
    processed_text = text.lower().replace(".", "")
    
    # Step 2: Split into chunks after processing
    return [processed_text[i:i+chunk_size] for i in range(0, len(processed_text), chunk_size)]

# Example usage
text = """"Climate change is one of the biggest challenges of our time.
It affects weather patterns, sea levels, and biodiversity.

Governments around the world are working on policies to reduce carbon emissions.
Renewable energy sources like solar and wind are becoming more popular.

Individuals can also make a difference by reducing waste, conserving energy, and supporting sustainable practices."""

chunks = late_chunking(text, 20)
print(chunks)
```

```python
#chromadb 
from pypdf import PdfReader

def extract_text_from_pdf(pdf_path):
    reader = PdfReader(pdf_path)
    text = ""
    for page in reader.pages:
        text += page.extract_text() + "\n"
    return text

text = extract_text_from_pdf("recipe.pdf")
print(text[:500])  # preview first 500 characters
```

```python
def chunk_text(text, chunk_size=512):
    chunks = [text[i:i+chunk_size] for i in range(0, len(text), chunk_size)]
    return chunks

chunks_256 = chunk_text(text, 256)
chunks_512 = chunk_text(text, 512)
chunks_1024 = chunk_text(text, 1024)

print("256-token chunks:", len(chunks_256))
print("512-token chunks:", len(chunks_512))
print("1024-token chunks:", len(chunks_1024))
```

```python
#generating embeddings
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("all-MiniLM-L6-v2")

embeddings_256 = model.encode(chunks_256)
embeddings_512 = model.encode(chunks_512)
embeddings_1024 = model.encode(chunks_1024)

print("Embedding dimension:", len(embeddings_256[0]))
```

```python
import chromadb
client = chromadb.Client()
print("ChromaDB is working!")
```

```python
# Create three collections
collection_256 = client.create_collection("pdf_corpus_256")
collection_512 = client.create_collection("pdf_corpus_512")
collection_1024 = client.create_collection("pdf_corpus_1024")

# Store 256-token chunks
for i, chunk in enumerate(chunks_256):
    collection_256.add(
        documents=[chunk],
        embeddings=[embeddings_256[i]],
        ids=[f"chunk256_{i}"]
    )

# Store 512-token chunks
for i, chunk in enumerate(chunks_512):
    collection_512.add(
        documents=[chunk],
        embeddings=[embeddings_512[i]],
        ids=[f"chunk512_{i}"]
    )

# Store 1024-token chunks
for i, chunk in enumerate(chunks_1024):
    collection_1024.add(
        documents=[chunk],
        embeddings=[embeddings_1024[i]],
        ids=[f"chunk1024_{i}"]
    )
```

```python
query = "What are the ingredients for chocolate cake?"

results_256 = collection_256.query(query_texts=[query], n_results=3)
results_512 = collection_512.query(query_texts=[query], n_results=3)
results_1024 = collection_1024.query(query_texts=[query], n_results=3)

print("Results from 256-token chunks:", results_256)
print("Results from 512-token chunks:", results_512)
print("Results from 1024-token chunks:", results_1024)
```
